```python
num_cols = []
for col in data.columns:
    # ÐšÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ Ð¿ÑƒÑÑ‚Ñ‹Ñ… Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ð¹ 
    temp_null_count = data[data[col].isnull()].shape[0]
    dt = str(data[col].dtype)
    if temp_null_count>0:
        num_cols.append(col)
#         temp_perc = round((temp_null_count / total_count) * 100.0, 2)
#         print('ÐšÐ¾Ð»Ð¾Ð½ÐºÐ° {}. Ð¢Ð¸Ð¿ Ð´Ð°Ð½Ð½Ñ‹Ñ… {}. ÐšÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ Ð¿ÑƒÑÑ‚Ñ‹Ñ… Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ð¹ {}, {}%.'.format(col, dt, temp_null_count, temp_perc))
```

```python
num_cols
```

```python
data_num = data[num_cols]
data_num
```

```python
from sklearn.impute import SimpleImputer
from sklearn.impute import MissingIndicator
```

```python
import sklearn
sklearn.__version__
```

```python
data_num_FuelEconomy = data_num[['FuelEconomy']]
data_num_FuelEconomy.head()
```

```python
indicator = MissingIndicator()
mask_missing_values_only = indicator.fit_transform(data_num_FuelEconomy)
mask_missing_values_only
```

```python
strategies=['mean', 'median','most_frequent']
```

```python
def test_num_impute(strategy_param):
    imp_num = SimpleImputer(strategy=strategy_param)
    data_num_imp = imp_num.fit_transform(data_num_FuelEconomy)
    return data_num_imp[mask_missing_values_only]
```

```python
for i in range(len(strategies)):
    print(strategies[i], test_num_impute(strategies[i]))
```
