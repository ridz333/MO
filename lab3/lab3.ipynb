## 1.2. "Внедрение значений" - импьютация (imputation)


```python
num_cols = []
for col in data.columns:
    # Количество пустых значений 
    temp_null_count = data[data[col].isnull()].shape[0]
    dt = str(data[col].dtype)
    if temp_null_count>0:
        num_cols.append(col)
#         temp_perc = round((temp_null_count / total_count) * 100.0, 2)
#         print('Колонка {}. Тип данных {}. Количество пустых значений {}, {}%.'.format(col, dt, temp_null_count, temp_perc))
```

```python
num_cols
```

```python
data_num = data[num_cols]
data_num
```

```python
from sklearn.impute import SimpleImputer
from sklearn.impute import MissingIndicator
```

```python
import sklearn
sklearn.__version__
```

```python
data_num_FuelEconomy = data_num[['FuelEconomy']]
data_num_FuelEconomy.head()
```

```python
indicator = MissingIndicator()
mask_missing_values_only = indicator.fit_transform(data_num_FuelEconomy)
mask_missing_values_only
```

```python
strategies=['mean', 'median','most_frequent']
```

```python
def test_num_impute(strategy_param):
    imp_num = SimpleImputer(strategy=strategy_param)
    data_num_imp = imp_num.fit_transform(data_num_FuelEconomy)
    return data_num_imp[mask_missing_values_only]
```

```python
for i in range(len(strategies)):
    print(strategies[i], test_num_impute(strategies[i]))
```

## 1.2.2. Обработка пропусков в категориальных данных

```python
# Выберем категориальные колонки с пропущенными значениями
# Цикл по колонкам датасета
cat_cols = []
for col in data.columns:
    # Количество пустых значений 
    temp_null_count = data[data[col].isnull()].shape[0]
    dt = str(data[col].dtype)
    if temp_null_count>0 and (dt=='object'):
        cat_cols.append(col)
        temp_perc = round((temp_null_count / total_count) * 100.0, 2)
        print('Колонка {}. Тип данных {}. Количество пустых значений {}, {}%.'.format(col, dt, temp_null_count, temp_perc))
```

## 3. Масштабирование данных
Термины "масштабирование" и "нормализация" часто используются как синонимы. Масштабирование предполагает изменение диапазона измерения величины, а нормализация - изменение распределения этой величины.

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, Normalizer
```

```python
sc1 = MinMaxScaler()
sc1_data = sc1.fit_transform(data[['Distance']])
```

```python
plt.hist(data['Distance'], 100)
plt.show()
```

```python
plt.hist(sc1_data, 50)
plt.show()
```

```python
sc2 = StandardScaler()
sc2_data = sc2.fit_transform(data[['Distance']])
```

```python
plt.hist(sc2_data, 50)
plt.show()
```
